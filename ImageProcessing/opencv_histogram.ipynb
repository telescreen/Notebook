```python
rainbow = cv2.imread(CFG.data_folder + 'rainbow.jpg')
show_rainbow = cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)
blue_bricks = cv2.imread(CFG.data_folder + '/bricks.jpg')
show_bricks = cv2.cvtColor(blue_bricks, cv2.COLOR_BGR2RGB)
horse = cv2.imread(CFG.data_folder + 'horse.jpg')
show_horse = cv2.cvtColor(horse, cv2.COLOR_BGR2RGB)
```

```python
# OpenCV BGR
hist_values = cv2.calcHist([blue_bricks], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
```

```python
def plot_histogram(img, color='bgr'):
    for i, col in enumerate(color):
        histr = cv2.calcHist([img], channels=[i], mask=None, histSize=[256], ranges=[0, 256])
        plt.plot(histr, color=col)
    plt.xlim(0, 256)
    plt.title('Histogram')
```

```python
plot_histogram(blue_bricks)
```

```python
plot_histogram(horse)
```
