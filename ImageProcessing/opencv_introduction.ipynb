# Draw multiple shapes on Image

### Draw shapes
We can draw shapes using `cv2.line`, `cv2.rectangle`, and `cv2.cicle`

```python
blank_img = np.zeros((512, 512, 3), dtype=np.int16)
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, (384, 10), (500, 150), color=(0, 255, 0), thickness=10)
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, (200, 200), (300, 300), color=(0,0,255), thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img, (100, 100), radius=50, color=(255,0,0), thickness=8)
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img, (400, 400), radius=50, color=(255,0,0), thickness=-1)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img, (0, 0), (512, 512), color=(100, 255, 255), thickness=2)
plt.imshow(blank_img)
```

### Draw text

```python
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(blank_img, text='Hello', org=(10, 500), 
            fontFace=font, fontScale=4, color=(255, 255, 255), 
            thickness=3, lineType=cv2.LINE_AA)
```

```python
plt.imshow(blank_img)
```

```python
blank_img = np.zeros((512,512,3), dtype=np.int32)
plt.imshow(blank_img)
```
