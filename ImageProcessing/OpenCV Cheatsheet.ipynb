```python
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
I = cv2.imread('D:\Data\standard_test_images\mandril_color.tif', cv2.IMREAD_UNCHANGED)
I = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)
I.shape
```

```python
h, w = I.shape[:2]
angle = 45.0
scale = 1.0
affine = cv2.getRotationMatrix2D((w/2, h/2), angle, scale)
O = cv2.warpAffine(I, affine, (w, h))
```
