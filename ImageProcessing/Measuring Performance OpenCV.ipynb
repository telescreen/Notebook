## Measuring Performance with OpenCV

Use **cv.getTickFrequency**

```python
e1 = cv2.getTickCount()
for i in range(5, 49, 2):
    img = cv2.medianBlur(img, i)
e2 = cv2.getTickCount()
t = (e2 - e1)/cv2.getTickFrequency()
print(t)
```

Many of the OpenCV fucntions are optimized using SSE2 or AVX, etc. OpenCV run opmitized code if it is enabled. Use **cv2.useOptimized()** to check if it is enabled.

```python
cv2.useOptimized()
```

```python
%timeit res = cv2.medianBlur(img, 49)
```

```python
cv2.setUseOptimized(False)
%timeit res = cv2.medianBlur(img, 49)
```
