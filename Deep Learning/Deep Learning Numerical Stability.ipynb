```python
%matplotlib inline

import d2l
from mxnet import nd, autograd

x = nd.arange(-8, 8, 0.1)
x.attach_grad()
with autograd.record():
    y = x.sigmoid()
y.backward()
d2l.plot(x, [y, x.grad], legend=['sigmoid', 'gradient'], xlim=[-7.5, 7.5])
```

```python
with autograd.record():
    y = x.relu()
y.backward()
d2l.plot(x, [y, x.grad], legend=['sigmoid', 'gradient'], xlim=[-7.5, 7.5])    
```

```python
M = nd.random.normal(shape=(4,4))
print('A single matrix: ', M)
for i in range(100):
    M = nd.dot(M, nd.random.normal(shape=(4,4)))
print('A single matrix after multiplying 100 matrices: ', M)
```
