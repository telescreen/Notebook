```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
q = beta(5, 5)
obs = q.rvs(2000)
grid = np.linspace(0.01, 0.99, 100)

fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(obs, bins=30, density=True)
ax.plot(grid, q.pdf(grid), 'k-', linewidth=2)
plt.show()
```

```python
from scipy.stats import linregress

x = np.random.randn(100)
y = 2 * x + np.random.normal(0, 0.4, 100)
fig, ax = plt.subplots(figsize=(10,6))

gradient, intercept, r_value, p_value, std_err = linregress(x, y)
yhat = x*gradient + intercept

ax.plot(x, y, 'o', label='data points', alpha=0.6)
ax.plot(x, yhat, 'r-', label='approx line')

ax.legend()
plt.show()
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm

def f(x, y):
    return np.cos(x**2 + y**2) / (1 + x**2 + y**2)

xgrid = np.linspace(-3, 3, 50)
ygrid = xgrid
x, y = np.meshgrid(xgrid, ygrid)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, f(x, y), rstride=2, cstride=2, 
                cmap=cm.jet, alpha=0.7, linewidth=0.25)
ax.set_zlim(-0.5, 1.0)
plt.show()
```

```python
def subplots():
    fig, ax = plt.subplots()
    for spine in ['left', 'bottom']:
        ax.spines[spine].set_position('zero')        
    for spine in ['right', 'top']:
        ax.spines[spine].set_color('none')
    ax.grid()
    return fig, ax

fig, ax = subplots()
x = np.linspace(-2, 10, 200)
y = np.sin(x)
ax.plot(x, y, 'r-', linewidth=2, label='sine function', alpha=0.6)
ax.legend(loc='lower right')
plt.show()
```

```python
x = np.linspace(0, 5, 100)
fig, ax = plt.subplots(figsize=(10, 6))

for theta in np.linspace(0, 2, 10):
    f = np.cos(np.pi*theta*x) * np.exp(-x)
    ax.plot(x, f)
    
plt.show()
```
